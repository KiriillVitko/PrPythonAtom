```python
class HashMap:
    
    class Entry:
        
        def __init__(self, key, value):
            #TODO store key and value as attributes
            self.key = key
            self.value = value
            
        def get_key(self):
            #TODO return key
            return self.key
            
        def get_value(self):
            #TODO return value
            return self.value
            
        def __eq__(self, other):
            #TODO реализовать функцию сравнения
            #Пояснения: self - текущий объект entry, other - какой-то другой объект
            #Какое условие на равенство объектов в ассоциативном массиве?
            #Если объекты равны, то метод должен возвращать True, иначе False
            return self.key == other.key
    
    def __init__(self, bucket_num = 64):
        #TODO Создать массив для бакетов
        self.list = []
        self.capacity = bucket_num
        self.cap = self.capacity
        self.list = [None for _ in range(self.capacity)]
        self.counter = 0
            
    def get(self, key, default_value=None):
        #TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        index = hash(key) % self.capacity
        if self.list[index] is None:
            return default_value
        while self.list[index]:
            if self.list[index].key == key:
                return self.list[index].value
            else:
                if index < self.cap - 1:
                    index += 1
                else:
                    return default_value
                
    def put(self, key, value):
        #TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет
        index = hash(key) % self.capacity
        if self.list[index] is None:
            self.list[index] = self.Entry(key,value)
            self.counter += 1
        else:
            if self.list[index].key == key:
                self.list[index].value = value
            else:
                while index < self.capacity and self.list[index] is not None:
                    if self.list[index].key == key:
                        self.list[index].value = value
                        return
                    else:
                        index += 1
                if self.list[index] is None:
                    self.list[index] = self.Entry(key,value)
                    self.counter += 1
                else:
                    self.list.append(self.Entry(key,value))
                    self.counter += 1
                    self.cap += 1
                    
    def __len__(self):
        #TODO Возвращает количество Entry в массиве
        return self.counter

    def _get_hash(self, key):
        #TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key) % self.capacity
    

    def _get_index(self, hash_value):
        #TODO По значению хеша вернуть индекс элемента в массиве
        return hash_value % self.capacity
    
    class V_Iterator:
        
        def __init__(self):
            self.values = []
            
        def get(self, index):
            return self.values[index]
        
        def put(self, value):
            self.values.append(value)
        
        def __iter__(self):
            self.index = 0
            return self
        
        def __next__(self): 
            if self.index < len(self.values):
                self.index += 1
                return self.get(self.index - 1)
            else:
                raise StopIteration
    
    def values(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор значений (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        self.v_iter = self.V_Iterator()
        for i in self.list:
            if i is not None:
                self.v_iter.put(i.get_value())
        return self.v_iter
        
    class K_Iterator:
        
        def __init__(self):
            self.keys = []
            
        def get(self, index):
            return self.keys[index]
        
        def put(self, key):
            self.keys.append(key)
        
        def __iter__(self):
            self.index = 0
            return self
        
        def __next__(self): 
            if self.index < len(self.keys):
                self.index += 1
                return self.get(self.index - 1)
            else:
                raise StopIteration
    
    def keys(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор ключей (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        self.k_iter = self.K_Iterator()
        for i in self.list:
            if i is not None:
                self.k_iter.put(i.get_key())
        return self.k_iter
        
    class I_Iterator:
        def __init__(self):
            self.items = []
                
        def get(self, index):
            return self.items[index]
    
        def put(self, item):
            self.items.append(item)
                
        def __iter__(self):
            self.index = 0
            return self

        def __next__(self):
            if self.index < len(self.items):
                self.index += 1
                return self.get(self.index - 1)
            else:
                raise StopIteration
    
    def items(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор пар ключ и значение (tuples)
        #(на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        self.i_iter = self.I_Iterator()
        for i in self.list:
            if i is not None:
                self.i_iter.put((i.key, i.value))
        return self.i_iter
        
    def __str__(self):
        #TODO Метод, вызываемый при вызове print, пофантазируйте, что можно написать
        line =""
        for i in self.list:
                if i is not None:
                    line += "(" + str(i.key) + "/" + str(i.value) + ")"
        return line
```
